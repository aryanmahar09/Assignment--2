```python
# 1Write a user-defined function that takes a number as input and returns:
# • "Positive Even"
# • "Positive Odd"
# • "Negative"
# • "Zero"
# Use proper conditional statements and explain your logic.

def f1():
    n=int(input("Enter a number: "))
    if n==0:
        print("The number is zero")
    elif n<0:
        print("the number is negative")
    elif n % 2 ==0:
        print("the number is even")
    else:
        print("the number is odd")
   
f1()
```

```python
#Write a function that accepts three numbers and returns the largest number without using built-in
#functions like max().
#Explain how your conditional structure ensures correctness.
def largest():
    a=int(input("enter first number:"))
    b=int(input("enter second number:"))
    c=int(input("enter third number:"))
    if a>b and a>c:
        print("a is largest")
    elif b>a and b>c:
        print("b is largest")
    else:
        print("c is largest")
largest()
#it checks all possibilities logically and ensure one number must be the largest
```

```python
# Write a function that accepts a student’s marks and returns the grade based on the following logic:
# • 90+ → A
# • 75–89 → B
# • 50–74 → C
# • Below 50 → Fail
# Use nested conditional statements and justify your approach.
def grade():
    g=int(input("enter your marks:"))
    if g >= 90:
        print("Grade A")
    elif g >75 and g <= 89:
        print("Grade B")
    elif g >50 and g <= 74:
        print("Grade C")
    else:
        print("Grade F")
grade()
```

```python
# 4. Write a function that takes age as input.
# • If age is negative, raise an error.
# • If age is below 18, return "Minor", otherwise return "Adult".
# Explain how conditional statements help in input validation
def age():
    a=int(input("enter your age:"))
    if a <= 0:
        raise ValueError("Age cannot be negative")
    elif a < 18:
        print("You are a minor")
    elif a < 65:
        print("You are an adult")
    else:
        print("You are a senior citizen")
age()
#conditional statements allow us to check the validity of the input and ensure that it meets certain criteria before proceeding with further logic. In this case, we check if the age is negative and raise an error if it is, ensuring that we only work with valid age values.
```

```python
# # 5. Write a function that accepts a list of integers and:
# • Counts even numbers
# • Ignores negative numbers
# • Stops processing if zero appears
# Use both loops and conditional statements inside the function
def count_even_numbers():
    numbers = [1, 2, -3, 4, 0, 5, 6]
    even_count = 0
    for num in numbers:
        if num < 0:
            continue  # Ignore negative numbers
        elif num == 0:
            break  # Stop processing if zero appears
        elif num % 2 == 0:
            even_count += 1  # Count even numbers
    print(f"Count of even numbers: {even_count}")
count_even_numbers()
```
